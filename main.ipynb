### 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution

```python
def log_transform(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return 0

for column in numerical_df.columns:
    
    log_for_plot = []
    log_for_plot = numerical_df[column].apply(log_transform)
    
    x_axis1 = np.arange(numerical_df[column].min(), numerical_df[column].max(), 0.01)
    mean1 = stats.mean(numerical_df[column])
    stdev1 = stats.stdev(numerical_df[column])
    
    x_axis2 = np.arange(log_for_plot.min(), log_for_plot.max(), 0.01)
    mean2 = stats.mean(log_for_plot)
    stdev2 = stats.stdev(log_for_plot)
    
    plt.figure(figsize=(10, 5))

    plt.subplot(1,2,1)
    sns.distplot(numerical_df[column], bins=15, hist=True)
    plt.plot(x_axis1, norm.pdf(x_axis1, mean1, stdev1), color = 'c')
    plt.title(f"distplot {column}")
    
    plt.subplot(1,2,2)
    sns.distplot(log_for_plot, bins=15, hist=True)
    plt.plot(x_axis2, norm.pdf(x_axis2, mean2, stdev2), color = 'c')
    plt.title(f"log transformed {column}")
    
    plt.tight_layout()
    plt.show()
    

```
